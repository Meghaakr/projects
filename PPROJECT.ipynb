```python
import pandas as pd
df=pd.read_csv('/content/Thyroid_Dataset.csv')
df
```

```python
df.isna().sum()
```

REMOVE NULL VALUES

```python
df['sex']=df['sex'].fillna(df['sex'].mode()[0])
```

```python
df.isna().sum()
```

LABEL ENCODING

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['sex']=le.fit_transform(df['sex'])
df['on_thyroxine']=le.fit_transform(df['on_thyroxine'])
df['query_on_thyroxine']=le.fit_transform(df['query_on_thyroxine'])
df['on_antithyroid_medication']=le.fit_transform(df['on_antithyroid_medication'])
df['sick']=le.fit_transform(df['sick'])
df['pregnant']=le.fit_transform(df['pregnant'])
df['thyroid_surgery']=le.fit_transform(df['thyroid_surgery'])
df['I131_treatment']=le.fit_transform(df['I131_treatment'])
df['query_hypothyroid']=le.fit_transform(df['query_hypothyroid'])
df['query_hyperthyroid']=le.fit_transform(df['query_hyperthyroid'])
df['lithium']=le.fit_transform(df['lithium'])
df['goitre']=le.fit_transform(df['goitre'])
df['tumor']=le.fit_transform(df['tumor'])
df['goitre']=le.fit_transform(df['goitre'])
df['hypopituitary']=le.fit_transform(df['hypopituitary'])
df['psych']=le.fit_transform(df['psych'])
df['referral_source']=le.fit_transform(df['referral_source'])
```

```python
df.dtypes
```

```python
X=df.iloc[:,:-1]
X
```

```python
y=df.iloc[:,-1]
y
```

OVERSAMPLING

```python
from imblearn.over_sampling import SMOTE
oversample=SMOTE()
Xo,yo=oversample.fit_resample(X,y)
```

```python
yo.value_counts()
```

TRAIN TEST SPLIT

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(Xo,yo,test_size=0.3,random_state=1)
print(X_train.shape)
print(X_test.shape)
```

STANDARD SCALER

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit(X_train)
X_train_new=sc.transform(X_train)
X_test_new=sc.transform(X_test)
```

ALGORTHMS

1.KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=60)
knn.fit(X_train_new,Y_train)
y_pred=knn.predict(X_test_new)
y_pred
```

```python
from sklearn.metrics import ConfusionMatrixDisplay,classification_report
print(classification_report(Y_test,y_pred))
print(ConfusionMatrixDisplay.from_predictions(Y_test,y_pred))
```

2.NAVE BASE

```python
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb.fit(X_train_new,Y_train)
y_pred_nv=nb.predict(X_test_new)
print(y_pred_nv)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay,classification_report
print(classification_report(Y_test,y_pred_nv))
print(ConfusionMatrixDisplay.from_predictions(Y_test,y_pred_nv))
```

3.SVM

```python
from sklearn.svm import SVC
sv=SVC(kernel='linear')
sv.fit(X_train_new,Y_train)
y_pred_sv=sv.predict(X_test_new)
y_pred_sv
```

```python
from sklearn.metrics import ConfusionMatrixDisplay,classification_report
print(classification_report(Y_test,y_pred_sv))
print(ConfusionMatrixDisplay.from_predictions(Y_test,y_pred_sv))
```

XG BOOSTING

```python
import numpy as np
from xgboost import XGBClassifier
from sklearn.model_selection import cross_val_score
```

```python
def cross_val(classifier,num_splits=10):
  model=classifier
  scores=cross_val_score(model, Xo,yo,cv=10)
  print(np.round(scores,2))
  print(scores.mean())
```

```python
cross_val(XGBClassifier(n_estimators=5))
```
