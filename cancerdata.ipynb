```python
import pandas as pd
df=pd.read_csv('/content/cancer_data.csv')
df
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
X=df.iloc[:,:-1].values
X
```

```python
Y=df.iloc[:,-1].values
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=1)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler()
sc.fit(X_train)
X_train_new=sc.transform(X_train)
X_train_new
```

```python
X_test_new=sc.transform(X_test)
X_test_new
```

```python
#knn
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train_new,Y_train)
Y_pred=knn.predict(X_test_new)
Y_pred
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,Y_pred)*100)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay,classification_report
print(classification_report(Y_test,Y_pred))
print(ConfusionMatrixDisplay.from_predictions(Y_test,Y_pred))
```

```python
#navebayes
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb.fit(X_train_new,Y_train)
Y_pred=nb.predict(X_test_new)
print(Y_pred)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,Y_pred)*100)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay,classification_report
print(classification_report(Y_test,Y_pred))
print(ConfusionMatrixDisplay.from_predictions(Y_test,Y_pred))
```

```python
from sklearn.svm import SVC
sv=SVC(kernel='linear')
sv.fit(X_train_new,Y_train)
Y_pred_sv=sv.predict(X_test_new)
Y_pred_sv
```

```python
sv.predict(sc.transform([[18.5,19.8,130.1,516,.05]]))
```

```python
from sklearn.metrics import classification_report,ConfusionMatrixDisplay
print(classification_report(Y_test,Y_pred_sv))
print(ConfusionMatrixDisplay.from_predictions(Y_test,Y_pred_sv))
```

```python
#poly_svm
from sklearn.svm import SVC
sv=SVC(kernel='poly')
sv.fit(X_train_new,Y_train)
Y_pred_sv=sv.predict(X_test_new)
Y_pred_sv
```

```python
from sklearn.metrics import classification_report,ConfusionMatrixDisplay
print(classification_report(Y_test,Y_pred_sv))
print(ConfusionMatrixDisplay.from_predictions(Y_test,Y_pred_sv))
```

```python
from sklearn.svm import SVC
sv=SVC()
sv.fit(X_train_new,Y_train)
Y_pred_sv=sv.predict(X_test_new)
Y_pred_sv
```

```python
from sklearn.metrics import classification_report,ConfusionMatrixDisplay
print(classification_report(Y_test,Y_pred_sv))
print(ConfusionMatrixDisplay.from_predictions(Y_test,Y_pred_sv))
```
